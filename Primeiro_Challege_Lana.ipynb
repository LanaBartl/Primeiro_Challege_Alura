### Importação dos dados



Nessa etapa estou importantando as bases de dados das 4 lojas que iremos trabalhar e analizar. E a importação das bibliotecas.

```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja
```

#Concatenação e criação de colunas



```python
loja['loja_id'] ='loja 1'
loja2['loja_id'] = 'loja 2'
loja3['loja_id'] = 'loja 3'
loja4['loja_id'] = 'loja 4'

loja = pd.concat([loja, loja2, loja3, loja4])
loja.head()
```

#1. Análise do faturamento



Foi criada uma lista denominada “lojas”, na qual foram adicionadas as variáveis correspondentes a cada loja analisada quanto ao faturamento.

Por meio de um laço de repetição for, aplicou-se, para cada loja presente na lista, o somatório da coluna Preço, com arredondamento para duas casas decimais, sendo os resultados exibidos por meio do comando print.

A partir dessa análise, constatou-se que a Loja 2 apresentou o maior faturamento, totalizando 1.488.459,06, e que a diferença entre o maior e o menor faturamento foi de 103.961,48.

```python
lojas = [loja, loja2, loja3, loja4]

for loja in lojas:
  faturamento = loja['Preço'].sum().round(2)
  print(faturamento)
```

```python
Lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
Faturamento_Total = [1534509.12, 1488459.06, 1464025.03, 1384497.58]

plt.figure(figsize=(7, 7))
plt.pie(
    Faturamento_Total,
    labels=Lojas,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Participação no faturamento total por loja')
plt.axis('equal')  # deixa o gráfico em formato circular
plt.show()
```

# 2. Vendas por Categoria

Em seguida, para cada loja, aplicou-se o método value_counts() sobre a coluna “Categoria do Produto”, com o objetivo de contabilizar a quantidade de vendas realizadas em cada categoria.

```python
for i, loja in enumerate(lojas, start=1):
  print(f"\n=== Loja {i} ===")

  venda_por_categoria = loja['Categoria do Produto'].value_counts()
  print(venda_por_categoria)
```
